````python
# 외부데이터 사용
# Make a requests
# requests: HTTP 요청을 쉽게 해줌
# 크롬에 www.naver.com 이라고 검색 -> 크롬에 www.naver.com을 요청
# 크롬이 주소를 숫자 데이터로 바꾸어줌 -> 숫자는 특정한 위치인 ip
# www.naver.com 데이터를 크롬이 받고 우리는 그 문서를 시각적자료로 변환되어 보는거임
import requests
````

````python
import random

my_numbers = {}

for i in range(1, 6):
    random_number = random.sample(range(1, 46), 6)
    my_numbers[f'{i}회차'] = random_number

print(my_numbers)



# 이번주 로또 데이터 가져오기
LOTTO_URL = 'https://www.dhlottery.co.kr/common.do'
params = {
    'method': 'getLottoNumber',
    'drwNo': 1126,
}

r = requests.get(LOTTO_URL, params=params)
#print(r.url)
lotto_dict = r.json()

print(lotto_dict)



# 가져온 데이터에서 로또번호 추출하기
lucky_number = []
bonus_number = lotto_dict['bnusNo']

for key, value in lotto_dict.items():
    if 'drwtNo' in key:
        lucky_number.append(value)

print(f'이번주 당첨번호 : {lucky_number}, BN: {bonus_number}')



# 랜덤 번호와 이번주 복권번호 비교하기
from pprint import pprint
pprint(my_numbers)

for my_number in my_numbers.values():
    my_number_set = set(my_number)
    lucky_number_set = set(lucky_number)

    match_count = len(my_number_set & lucky_number_set)
    if match_count == 6:
        print('1등! 10억')
    elif match_count == 5 and (bonus_number in my_number):
        print('2등! 2천만원')
    elif match_count == 5:
        print('3등! 3백만원')
    elif match_count == 4:
        print('4등! 4만원' )
    elif match_count == 3:
        print('5등! 5천원')
    else:
        print('꽝')
````
